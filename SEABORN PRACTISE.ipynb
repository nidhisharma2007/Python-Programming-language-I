```python
import seaborn as sns
```

```python
x=sns.load_dataset('tips')
x
```

```python
#Scatterplot
sns.scatterplot(data=x,x="total_bill",y="tip")
#sns.scatterplot(data=x,x="total_bill",y="tip",hue='total_bill',pallete='Set2') 
#hue can work as legend and also style for differentiation
#sns.set_style('darkgrid') sets the background
#sns.set_context('notebook')
```

```python
#Histplot
sns.histplot(data=x,x='total_bill')
```

```python
sns.lineplot(data=x,x="total_bill",y="tip")
```

```python
sns.barplot(data=x,x="sex",y="tip")
```

```python
sns.countplot(data=x,x='sex')
```

```python
sns.boxplot(data=x,x='sex',y='tip')
```

```python
#violinplot
#Width of the violin:Density of the data at different values
#middle line:Median
#Thick bar:inner quartile range
sns.violinplot(data=x,x='sex',y='tip')
```

```python
#Kernal Density Estimation:for estimating the probability density function of continuous  or non-parametric data
#Simply a smooth curve that shows the distribution of data -like  a smoothed out version
#It helps you see where the data is concentrated(peaks) and how it spreads
sns.kdeplot(data=x,x='total_bill')
```

```python
#Heatmap
#2-D color-coded  matrix that shows relationships or patterns betweeen two variables
#Used to show correlation features
#Visualize confusion matrix

corr = x[['total_bill', 'tip', 'size']].corr()
sns.heatmap(corr, annot=True,cmap='Blues') #annotates all the correlations
```

```python
#pair plot=Multiple plots and visualize pair wise relationship
sns.pairplot(x[['total_bill', 'tip', 'size']])
```

```python
sns.relplot(
    data=x,
    x="total_bill",
    y="tip",
    hue="sex",
    kind="scatter"
)

```
