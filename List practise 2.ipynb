```python
#Data type
sports=['Basketball','kabadi','football','hockey','cricket','baseball','badminton','running','chess','tennis']
print(type(sports))
```

```python
#Indexing
sports=['Basketball','kabadi','football','hockey','cricket','baseball','badminton','running','chess','tennis']
print(sports[2:5])
```

```python
print(sports[ :9])
```

```python
print(sports[-9:-4])
```

```python
#Append Items
sports=['Basketball','kabadi','football','hockey','cricket','baseball','badminton','running','chess','tennis']
sports.append('Freefire')
print(sports)
```
