```python
# 1-Program to print even numbers between 1 and 50
# Skip multiples of 10 and stop if number > 40

for i in range(1, 51):
    if i % 2 == 0: # Check if number is even
        if i % 10 == 0: # Check if number is a multiple of 10
            continue
        elif i > 40: # Check if number is greater than 40
         break
        else:
            print(i)
    else:
        continue 
```

```python
#2A-Palindrome

num=int(input(" Enter any number:"))
rev=0
x=num
while(num>0):
  rev=(rev*10)+num%10
  num=num//10
if(x==rev):
  print("Given Number is Palindrome")
else:
  print("Given number is not Palindrome")
```

```python
#2B-Armstrong

n=int(input("Enter number:"))
x=n
sum=0
while(n>0):
  sum=sum+(n%10)*(n%10)*(n%10)
  n=n//10
if x==sum:

    
  print("Number is Armstrong")
else:
  print("Number is not Armstrong")
```

```python
#2C-Fibonnaci series

n=int(input("Enter the limit number"))
x=0
y=1
z=0
while(z<=n):
  print(z)
  x=y
  y=z
  z=x+y
```

```python
#3-Finding factorial using function

def factorial(number):
    if number == 0 or number == 1:
        return 1
    else:
        return number * factorial(number - 1)
```

```python
#calling the function
factorial(5)
```
