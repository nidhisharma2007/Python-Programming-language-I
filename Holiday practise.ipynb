```python
x=[1,2,3,4,5]
f=[2,4,6,8,10]
fx=[]
for n in range(len(x)):
  prod=x[n]*f[n]
  fx.append(prod)
  sum_fx=sum(fx)
  sum_f=sum(f)
  mean=sum_fx/sum_f
print(mean)
```

```python
x=[2,3,4,None,5,6,7]
sum=0
count=0
for i in x:
    if i is not None:
        sum+=i
        count+=1
        mean=sum/count
imputed=[]
for i in x:
    if i is None:
        imputed.append(mean)
    else:
        imputed.append(i)

print('Original data:', x)
print('Imputed data:',imputed)
```

```python
data = [2, 5, 5, 6, 6, 7]
freq = {}  

for x in data:
    freq[x] = freq.get(x, 0) + 1

max_freq = max(freq.values())
modes = []

for k, v in freq.items():
    if v == max_freq:
        modes.append(k)

print("dataset mode:", modes)
```

```python
dataset = [48, 49, 50, 51, 52]
n = len(dataset)
dataset_mean = sum(dataset) / len(dataset)
sum_sqrd = 0  
for x in dataset:
  sum_sqrd += (x - dataset_mean) ** 2
std_dev = (sum_sqrd / len(dataset)) ** 0.5
print(std_dev)
```
