```python
## Take user input for two numbers
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# Perform arithmetic operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
exponentiation = num1 ** num2
division = num1 / num2 
modulus = num1 % num2 
floor_division = num1 // num2 

    
# Display results with appropriate messages
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication )
print("Division:", division )
print("Modulus:" , modulus )
print("Exponentiation:" , exponentiation )
print("Floor Division:" , floor_division)
```

```python
# Original list
my_list = [1, 2, 3, 4, 5]
print("Original list:", my_list)

#Swapping using temperory variable

i1=0
i2=3

temp = my_list[i1]      
my_list[i1] = my_list[i2]  
my_list[i2] = temp

print('Swapped list:',my_list)

#Swappimng without using a temperory variable
my_list = [1, 2, 3, 4, 5]
my_list[i1], my_list[i2] = my_list[i2], my_list[i1]

print('swapped list:',my_list)

```

```python

my_dict = {'name': 'Alice', 'age': 30, 'city': 'New York'}
print("Initial dictionary:", my_dict)

# Add elements
my_dict['job'] = 'Engineer' 
print('UPDATED:',my_dict)
my_dict.update({'salary': 75000}) #update method
print("updated dictionary:", my_dict)

# Remove elements
del my_dict['age'] #delete method
print("removed item:", my_dict)
removed_item = my_dict.pop('city')  #pop method
print("Removed (city):", removed_item)

# Copy item
dict_copy = my_dict.copy() #copy methopd
print("Copied dictionary:", dict_copy)
copied_job = dict(my_dict) #dict  method
print('COPIED DICTIONARY:',copied_job)

# Create nested dictionaries
my_dict2 = {
  "address" : {
    "street" : "123 Main St",
    "zip" : 10001
  },
  "person" : {
    "name" : "my_dicT2",
    "year" : 2024 }
  }
print(my_dict2)
```
