```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#basic graph
x=[2023,2024,2025,2026]
y=[15,25,30,20] #Class size
plt.plot(x,y) #coordinates of x and y axis
plt.show()
```

```python
#This will be only for one arguement for example for y axis so x axis arguements are given by default
x=[2023,2024,2025,2026]
y=[15,25,30,20] #Class size
plt.plot(y) #coordinates of x and y axis
plt.show()
```

```python
#numpy arrays are faster than python lists and gives more functionalities
x=np.array([2023,2024,2025,2026])
y=np.array([15,25,30,20])
plt.plot(x,y) 
plt.show()
```

```python
#Customization
x=np.array([2023,2024,2025,2026])
y=np.array([15,25,30,20])
plt.plot(x,y,marker=".") #Dot or period
plt.show()
```

```python
x=np.array([2023,2024,2025,2026])
y=np.array([15,25,30,20])
plt.plot(x,y,marker="o",markersize=20,markerfacecolor='Red',markeredgecolor='Green',linestyle='dashed',linewidth=3,color='Pink') 
plt.show()
```

```python
#inestyle -default is solid
#Color= for changiing line color
x=np.array([2023,2024,2025,2026])
y=np.array([15,25,30,20])
plt.plot(x,y,marker="v",linestyle="dotted",color='Green') #Down triangle
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r') #O bullet ,: dotted ,r:red
plt.show()
```

```python
x=np.array([2023,2024,2025,2026])
y=np.array([15,25,30,20])
plt.plot(x,y,marker="*",linestyle='none') 
plt.show()
```

```python
x=np.array([2023,2024,2025,2026])
y1=np.array([15,25,30,20])
y2=np.array([17,23,38,5])
line_style=dict(marker="*") #creating a dictionary
plt.title('Class size',fontsize=20,family='Arial',fontweight="bold",color='Blue')
plt.xlabel('Year',fontsize=15,family='Arial',fontweight="bold",color='green')
plt.ylabel('Students',fontsize=15,family='Arial',fontweight="bold",color='green')
plt.plot(x,y1,** line_style) #unpacking the dictionary
plt.plot(x,y2,** line_style)
plt.show()
```

```python
x=np.array([2023,2024,2025,2026])
y1=np.array([15,25,30,20])
y2=np.array([17,23,38,5])
plt.plot(x,y1) 
plt.plot(x,y2) 
plt.title('Class size',fontsize=20,family='Arial',fontweight="bold",color='Blue')
plt.xlabel('Year',fontsize=15,family='Arial',fontweight="bold",color='green')
plt.ylabel('Students',fontsize=15,family='Arial',fontweight="bold",color='green')

plt.show()
```

```python
#grid() helps make plots easier to read by adding reference lines
x=np.array([2023,2024,2025,2026])
y1=np.array([15,25,30,20])
plt.grid()
plt.plot(x,y1) 
plt.show()
```

```python
#bar charts
#compares categories of data by representing each category with a bar
categories=['Grains','Fruit','Vegetable','Protein','Dairy','Sweets']
values=[4,3,2,5,3,1]
plt.bar(categories,values,color='Green')
plt.title('Daily Consumption')
plt.xlabel('Food')
plt.ylabel('Quantity')
        
plt.show()
```

```python
#Pie charts
#Circular chart divided into slices to show percentages of the total
#Good for visualising distribution among the categories
categories=['Freshmen','Softmores','Juniors','Seniors']
values=np.array([300,250,275,225])
colors=['Blue','Red','Green','Pink']
plt.pie(values,labels=categories,colors=colors)
plt.legend()
plt.title('College')
plt.show()
```

```python
#Scattergraph
#Shows the correlation between two variables
#Helps to identify a correlation(+,-,None)
x=[0,1,1,2,3,4,5,6,7,7,8] #Hours studied
y=[55,60,65,62,68,70,75,78,82,85,87] #Test scores
plt.scatter(x,y,color='green',s=100)
plt.title('Grades')
plt.xlabel('Hours studied')
plt.ylabel('Test scores')
plt.show()
```

```python
#Histograms
#A visual representation of the distribution of quantitative data
#They group values into bins(intervals) and counts how many fall in each range
#good for statisr=tical probability
scores=np.random.normal(170,250,700)#mean,std_dev,size #generates random data that follows normal gaussian distribution
plt.hist(scores,bins=10)
plt.show()


```

```python
#subplots
#figure=The entire canvas
#Axes(Ax)=a single plot(subplot)
#Subplot allows you to show multiple graphs in one single figure

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) #rows,columns,index
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```
